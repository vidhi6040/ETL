```python
data["transaction_id"] = data["transaction_id"].str.strip()
```

```python
cleaned_data = data[data["transaction_id"].notna() & (data["transaction_id"] != "") & (data["transaction_id"] != "nan")]
cleaned_data
```

```python
empty_data = data[data["transaction_id"].isna() | (data["transaction_id"] == "") | (data["transaction_id"] == "nan")]
empty_data
```

```python
cleaned_data.to_csv("transactions_cleaned.csv", index=False)
empty_data.to_csv("transactions_empty.csv", index=False)
```
