```python
import pandas as pd
```

```python
df_cleaned = pd.read_csv("C:\\Users\\mailt\\Desktop\\Diya\\Projects\\ETL\\transactions_cleaned.csv")
```

```python
df_cleaned["transaction_date"] = pd.to_datetime(df_cleaned["transaction_date"], format="%d/%m/%Y", errors="coerce")
```

```python
df_cleaned["month"] = ( df_cleaned["transaction_date"].dt.to_period("M").astype(str))
```

```python
monthly_sales = ( 
        df_cleaned
        .groupby("month", dropna=True)
        .agg(
            total_amount=("amount", "sum"),
            total_transactions=("transaction_id", "count")
        )
        .reset_index()
    )
```

```python
monthly_sales.to_csv("monthly_sales.csv", index=False)
```
